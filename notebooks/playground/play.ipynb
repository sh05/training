```python
%matplotlib inline
import matplotlib.pyplot as plt
import japanize_matplotlib
import MeCab, re, json, glob
from sklearn.decomposition import PCA
from gensim import models
m = MeCab.Tagger("-Owakati")
```

```python
def get_posts(dirname):
    posts = []
    files = glob.glob('./dumps/channel/{}/*.json'.format(dirname))
    for file in files:
        with open(file) as f:
            posts += json.loads(f.read())
    return posts
```

```python
p_dic = {
    'yutakamiyaji': [],
    'shogo': [],
    'marutaku': [],
    'riku': [],
}

b_posts = get_posts('bookmarks')
f_posts = get_posts('fyi')
        
for p in b_posts:
    if "username" not in p or p["username"] != "IFTTT":
        continue
    for a in p["attachments"]:
        try:
            p_dic[a["text"]].append(a["title"])
        except:
            pass
        
for p in f_posts:
    if "username" not in p or p["username"] != "IFTTT":
        continue
    for a in p["attachments"]:
        try:
            p_dic[a["title"]].append(a["text"])
        except:
            pass
        
users = p_dic.keys()
print([[u, len(p_dic[u])] for u in users])
```

```python
_tag = re.compile(r'<.*?>')
_url = re.compile(r'(http|https)://([-\w]+\.)+[-\w]+(/[-\w./?%&=]*)?')
_title = re.compile(r'( - ).*$')
_par = re.compile(r'\(.*?\)')
_sla = re.compile(r'/.*$')
_qt = re.compile(r'"')
_sep = re.compile(r'\|.*$')
_twi = re.compile(r'(.*)on Twitter: ')
_lab = re.compile(r'(.*) ⇒ \(')
_lab2 = re.compile(r'(.*)&amp;gt;\(')
_last_par = re.compile(r'(\))$')
_line = re.compile('[\r\n]+$')
def com_clean_text(text):
    text = text.translate(str.maketrans({chr(0xFF01 + i): chr(0x21 + i) for i in range(94)}))
    text = re.sub(_lab, '', text)
    text = re.sub(_lab2, '', text)
    text = re.sub(_tag, '', text)
    text = re.sub(_url, '', text)
    text = re.sub(_title, '', text)
    text = re.sub(_sla,  '', text)
    text = re.sub(_qt,  '', text)
    text = re.sub(_sep, '', text)
    text = re.sub(_twi, '', text)
    text = re.sub(_par, '', text)
    text = re.sub(_last_par, '', text)
    text = re.sub(_line, '', text)
    return text
```

```python
def get_not_oneletter(text):
    return ' '.join([t for t in text.split(' ') if len(t) > 1])
```

```python
%%time
p_all = []
m_all = []
for u in users:
    p_dic[u] = list(map(com_clean_text, p_dic[u]))
    tmp = list(map(get_not_oneletter, [m.parse(p).split('\n')[0] for p in p_dic[u]]))
    m_all += [wakati.split(' ') for wakati in tmp]
    p_all += [u + '**' + p for p in p_dic[u]]
```

```python
# 参考記事： http://qiita.com/okappy/items/32a7ba7eddf8203c9fa1
class LabeledListSentence(object):
    def __init__(self, words_list, labels):
        self.words_list = words_list
        self.labels = labels

    def __iter__(self):
        for i, words in enumerate(self.words_list):
            yield models.doc2vec.TaggedDocument(words, ['%s' % self.labels[i]])

# gensim にクチコミを登録
# クチコミに会社名を付与するため、参考記事で実装されていた拡張クラスを使っています
sentences = LabeledListSentence(m_all, p_all)

# doc2vec の学習条件設定
# alpha: 学習率 / min_count: X回未満しか出てこない単語は無視
# size: ベクトルの次元数 / iter: 反復回数 / workers: 並列実行数
model = models.Doc2Vec(
    window=2,
    dm=1,
    alpha=0.025,
    min_count=2,
    vector_size=200,
    epoch=50,
    workers=4
)

# doc2vec の学習前準備(単語リスト構築)
model.build_vocab(sentences)

# 学習実行
model.train(
    sentences,
    total_examples=len(m_all),
    epochs=model.epochs
)

# セーブ
model.save('./doc2vec.model')

# 学習後はモデルをファイルからロード可能
# model = models.Doc2Vec.load('./data/doc2vec.model')

# 順番が変わってしまうことがあるので会社リストは学習後に再呼び出し
labels = model.docvecs.offset2doctag
```

```python
def w_draw_scatter_plot(vecs, tags):
    # Scikit-learnのPCAによる次元削減とその可視化
    pca = PCA(n_components=2)
    coords = pca.fit_transform(vecs)
    
    tsne = TSNE(n_components=2)
    coords = tsne.fit_transform(vecs)

    # matplotlibによる可視化
    fig, ax = plt.subplots(figsize=(16, 16))
    x = [v[0] for v in coords]
    y = [v[1] for v in coords]

    ax.scatter(x, y, c='red', s=30, alpha=0.5, linewidth=0.2, edgecolors='#777777')
        
    area_dic = {
    "right" : sum([np.where(x == v)[0].tolist() for v in x if v > 0.015], []), 
    "left" : sum([np.where(x == v)[0].tolist() for v in x if v < -1 * 0.005], []),
    "top" : sum([np.where(y == v)[0].tolist() for v in y if v > 0.005], []),
    "bottom" : sum([np.where(y == v)[0].tolist() for v in y if v < -1 * 0.005], []),
    "center" : sum([np.where(coords == c)[0].tolist() for c in coords if c[0]**2 + c[1]**2 < 0.0002 ** 2], [])
    }
    
    area_dic = {
    "right" : sum([np.where(x == v)[0].tolist() for v in x if v > 40], []), 
    "left" : sum([np.where(x == v)[0].tolist() for v in x if v < -1 * 40], []),
    "top" : sum([np.where(y == v)[0].tolist() for v in y if v > 40], []),
    "bottom" : sum([np.where(y == v)[0].tolist() for v in y if v < -1 * 40], []),
    "center" : sum([np.where(coords == c)[0].tolist() for c in coords if c[0]**2 + c[1]**2 < 5 ** 2], [])
    }
    
    for area in area_dic.keys():
        for i in area_dic[area]:
            ax.annotate(tags[i], (coords[i][0], coords[i][1]), fontproperties=fp)
        area_dic[area] = [tags[i] for i in area_dic[area]]
    
                
    plt.tight_layout()
#     plt.show()
#     plt.savefig('./bookmarks.jpg')
    return area_dic
```
