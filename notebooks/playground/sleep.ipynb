```python
import requests
```

```python
url = "http://api.e-stat.go.jp/rest/2.1/app/getStatsData?appId=2f941e0daa0d086855f2207b3a4103c660fe1fe5&lang=J&statsDataId=0003224280&metaGetFlg=Y&cntGetFlg=N&sectionHeaderFlg=1"
response = requests.get(url)
print(response)
```

```python
print(response.headers['Content-Type'])
```

```python
xmldata = response.text
```

```python
import xml.etree.ElementTree as ET

root = ET.fromstring(xmldata)
```

```python
print(root)
print(root.tag)
print(root.attrib)
```
